```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#District Summary

total_schools = school_data_complete['school_name'].nunique()

total_students = school_data_complete['student_name'].count()

total_budget = school_data_complete['budget'].unique().sum()

avg_math_score = school_data_complete['math_score'].mean()

avg_reading_score = school_data_complete['reading_score'].mean()

passing_rate = (avg_math_score + avg_reading_score)/2

passing_math_count = school_data_complete[school_data_complete['math_score'] >= 70]['math_score'].dropna().count()
passing_math = passing_math_count / school_data_complete['math_score'].count()

passing_reading_count = school_data_complete[school_data_complete['reading_score'] >= 70]['reading_score'].dropna().count()
passing_reading = passing_reading_count / school_data_complete['reading_score'].count()


summary_df = pd.DataFrame([
    {"Total Schools" : total_schools, "Total Students" : total_students, 
     "Total Budget" : total_budget, "Average Math Score" : avg_math_score, 
    "Average Reading Score" : avg_reading_score,"% Passing Math" : passing_math,
    "Passing Reading" : passing_reading, "% Overall Passing" : passing_rate}
])

summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#School Summary
school_summary = school_data_complete.groupby(['school_name']).agg({
    "reading_score" : ["mean"], "math_score" : ["mean"],
})

school_summary['Total Students'] = school_data_complete['school_name'].value_counts()

school_types = school_data.set_index(['school_name'])['type']
school_summary['school_types'] = school_types

school_budget = school_data.set_index(['school_name'])['budget']
school_summary['school_budget'] = school_budget
school_summary['per_student_budget'] = school_summary['school_budget'] / school_summary['Total Students']


df = school_data_complete.set_index(['school_name'])
df_index = school_data_complete['school_name'].unique()

#calculates number of students passing math
math_numbers = []
for x in df_index:
    
     df_slice = df.loc[x] ['math_score'] >=70
     m_num = (df_slice.to_frame()['math_score'].sum())
     list_num = [x, m_num]
     math_numbers.append(list_num)
    
percent_math_school = pd.DataFrame(math_numbers, columns= ['school_name', 'number_passing_math'])

#calculates number of students passing reading
reading_numbers = []
for x in df_index:
    
     df_slice = df.loc[x] ['reading_score'] >=70
     m_num = (df_slice.to_frame()['reading_score'].sum())
     list_num = [x, m_num]
     reading_numbers.append(list_num)

percent_reading_school = pd.DataFrame(reading_numbers, columns= ['school_name', 'number_passing_reading'])

#merges number passing reading and math
percent_passing_df = pd.merge(percent_reading_school, percent_math_school, how="left", on=["school_name", "school_name"])

#merges number passing and school summary
school_summary_df = pd.merge(school_summary, percent_passing_df, how="left", on=["school_name", "school_name"])

#calculates percentage passing reading, math, overall
school_summary_df['% passing reading'] = \
(school_summary_df['number_passing_reading'] / school_summary_df.iloc[:, 3]) * 100


school_summary_df['% passing math'] = \
(school_summary_df['number_passing_math'] / school_summary_df.iloc[:, 3]) *100


school_summary_df['% overall passing'] = \
(school_summary_df['% passing reading'] + school_summary_df['% passing math']) / 2

#drops unnecessary columns and cleans column names
school_summary_df = school_summary_df.drop(['number_passing_reading', 'number_passing_math'], 1)

summary_columns = ["School Name", "Average Reading Score", "Average Math Score", "Total Students", "School Type", 
                   "Total School Budget", "Per Student Budget", 
                   "% Passing Reading", "% Passing Math", "% Overall Passing"]

school_summary_df.columns = summary_columns


#indexes on school name
indexed_school_summary = school_summary_df.set_index(['School Name'])
indexed_school_summary
```
